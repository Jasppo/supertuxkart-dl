# Videos of matches

```python
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir("data/") if isfile(join("data/", f))]
```

```python
import re
import numpy as np
import matplotlib.pyplot as plt

def Filter(string, substr):
    return [k for k in string if substr in k]

def match_video(match_number = 1, player_number = 0):
    # Player 0
    images = []
    coords = []
    substr = 'M0' + str(match_number) + '_player0' + str(player_number)
    player = Filter(onlyfiles, substr)
    csv_names = sorted(Filter(player, '.csv'))
    image_names = sorted(Filter(player, '.png'))
    from matplotlib.image import imread
    
    for csv_name in csv_names:        
        csv_path = 'data/' + csv_name
        coords.append(np.loadtxt(csv_path, dtype=np.float32, delimiter=','))
        
    for image_name in image_names:
        image_path = 'data/' + image_name
        images.append(imread(image_path))
        
    image_data = []
        
    for idx, image in enumerate(images):
        print(coords[idx])
        fig = plt.figure()
        ax = fig.add_axes([0., 0., 1., 1.])
        ax.clear()
        ax.imshow(image)
        WH2 = np.array([400, 300])/2
        ax.add_artist(plt.Circle(WH2*(coords[idx] + 1), 2, ec = 'r', fill = False, lw = 15))
        ax.axis('off')
        fig.canvas.draw()
        X = np.array(fig.canvas.renderer._renderer)
        
        plt.show()
        plt.close()
        
        image_data.append(X)
        
    return image_data
        
```

```python
def show_video(frames, fps = 30):
    import imageio
    from IPython.display import Video, display
    
    imageio.mimwrite('test.mp4', frames, fps=fps, bitrate=10000000)
    display(Video('test.mp4', width = 800, height = 600, embed = False))
    
```

# Percentage of image data where the puck is not in the image

```python
from os import listdir
from os.path import isfile, join
import numpy as np
onlyfiles = [f for f in listdir("data/") if isfile(join("data/", f))]

def Filter(string, substr):
    return [k for k in string if substr in k]

csv_names = Filter(onlyfiles, '.csv')

total = 0
not_in_image = 0

for csv_name in csv_names:
    total += 1
    csv_path = 'data/' + csv_name
    label = np.loadtxt(csv_path, dtype=np.float32, delimiter=',')
    if label[0] == 0. and label[1] == 1.:
        not_in_image += 1

print(f"Total images: {total}, # of images without puck: {not_in_image}, Percentage: {not_in_image/total}")
```
